```python

from sklearn import datasets
import pandas as pd

print("This is the program to access iris dataset")
iris = datasets.load_iris()

print("To print the description of iris dataset")
print(iris.DESCR)

print("\n\n\n")

# df will hold dataset as a table
df = pd.DataFrame(
    iris.data,
    columns=iris.feature_names
)

# labels are assigned to df (target)
df['target'] = pd.Series(iris.target)

df['target_names'] = df['target'].apply(
    lambda y: iris.target_names[y]
)

print("To display first 5 samples")
# df.head(5) will return the 1st 5 samples of dataset
print(df.head(5))


print("To display 5 samples randomly")
print(df.sample(5))

# Train test split ratio
from sklearn.model_selection import train_test_split

df_train, df_test = train_test_split(df, test_size=0.3)  # for 70:30 split

print("The total no. of samples in the dataset =", df.shape[0])
print("The no. of samples in training set =", df_train.shape[0])
print("The no. of samples in testing set =", df_test.shape[0])

print("The first five samples of train")
print(df_train.head(5))

print("The first five samples of test")
print(df_test.head(5))
```
